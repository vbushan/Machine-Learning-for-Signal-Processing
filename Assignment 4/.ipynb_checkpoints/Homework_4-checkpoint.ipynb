#### Loading the MFCC matrix:

```python
mfcc_mat = loadmat('mfcc.mat')
mfcc_matX = mfcc_mat['X']
```

```python
mfcc_matX.shape
```

#### It can be seen that there is a X matrix, which holds 962 MFCC vectors, each of which has 12 coefficients.

#### Loading the mean and covariance matrices from MuSigma.mat:

```python
muSigma = loadmat('MuSigma.mat')
```

#### Loading the matrix mX from it:

```python
mX = muSigma['mX']
mX.shape
```

#### In the mX matrix, first column vector contains the mean vector of MFCCs of piano-only frames. Whereas, the second column vector holds the mean vector of MFCCs of the claps.

#### Loading the matrix sigma from it:

```python
sigma = muSigma['Sigma']
sigma.shape
```

#### Sigma is a 12x12x2 3-D array in which the first 2D slice(12x12) is a covariance matrix of the piano part and the upper 2D slide is for the clap sound. Separating the mean and covariances for the two sounds:

```python
mX_piano = mX[:,0]
mX_clap = mX[:,1]
sigma_piano = sigma[:,:,0]
sigma_clap = sigma[:,:,1]
print(sigma_clap.shape)
print(sigma_piano.shape)
```

#### Calculating the p.d.f. of an MFCC vector in X for the two multivariate normal (Gaussian) distribution you can define from the two sets of means and cov matrices:

```python
pdf_piano = multivariate_normal.pdf(mfcc_matX.T, mean = mX_piano, cov = sigma_piano)
pdf_clap = multivariate_normal.pdf(mfcc_matX.T, mean = mX_clap, cov = sigma_clap)
print(pdf_piano.shape)
print(pdf_clap.shape)
```

#### Stacking the two PDFs together to form the 2x962 matrix P:

```python
P = np.vstack((pdf_piano,pdf_clap))
P.shape
```

#### Normalizing the P matrix:

```python
P_norm_1 = np.zeros((962,))
P_norm_2 = np.zeros((962,))
for i in range(P.shape[1]):
    P_norm_1[i] = np.true_divide(P[0,i],P[0,i] + P[1,i])
    P_norm_2[i] = np.true_divide(P[1,i],P[0,i] + P[1,i])
P_norm = np.vstack((P_norm_1,P_norm_2))
print(P_norm.shape)
```

```python
print(P_norm)
```

#### Plotting the matrix as an image:

```python
sns.heatmap(P_norm)
```

#### Although we can observe the wrong claps in the middle. But, we don't like the result and would like to come up with a more smooth result.. In order to do that, we would want a transition matrix with some dominant diagonal elements.

```python
T = np.array([[0.9,0.1],[0,1]])
```

#### Applying this matrix to the normalized P matrix defined above:

#### Initializing the first column vector of the new posterior probability matrix:

```python
post_prob_P = np.zeros((2,962))
print(post_prob_P.shape)
post_prob_P[:,0] = P_norm[:,0]
```

#### Now we will define a class index 'b' to pick up the corresponding transition probabilities and then multiply the transition probabilities with the existing posterior in a recursive way:

```python
n_iter = 100
for x in range(n_iter):
    for t in range(P_norm.shape[1] -1):
        b = np.argmax(post_prob_P[:,t])
        post_prob_P[:,t+1] = np.multiply(np.transpose(T[b,:]), P_norm[:,t+1])
print(post_prob_P.shape)
```

#### Normalising the posterior probability matrix:

```python
Post_norm_1 = np.zeros((962,))
Post_norm_2 = np.zeros((962,))
for i in range(post_prob_P.shape[1]):
    Post_norm_1[i] = np.true_divide(post_prob_P[0,i],post_prob_P[0,i] + post_prob_P[1,i])
    Post_norm_2[i] = np.true_divide(post_prob_P[1,i],post_prob_P[0,i] + post_prob_P[1,i])
Post_norm = np.vstack((Post_norm_1,Post_norm_2))
print(Post_norm.shape)
```

#### Plotting the above matrix using heatmap:

```python
sns.heatmap(Post_norm)
```

#### To get better smoothing, we will implement Viterbi Algorithm:

#### We will calculate the smoothed posterior probability matrix in a similiar way. Firstly, we will initialized the first frame of the matrix similiarly.

```python
viterbi_post = np.zeros((2,962))
viterbi_post[:,0] = P_norm[:,0]
```

 #### This time we check on all paths to pick up the best one by calculating all the probabilities of state transitions:

```python
B = np.zeros((2,962))
for t in range(viterbi_post.shape[1] -1):
    vit_b_c1 = np.argmax(T[:,0]*viterbi_post[:,t])
    B[0,t+1] = vit_b_c1
    viterbi_post[0,t+1] = T[vit_b_c1,0]*viterbi_post[vit_b_c1,t]*P[0,t+1]
print(viterbi_post.shape)
```

#### We got the above posterior probabilities for Piano.

#### Repeating the above procedure to obtain the posterior probabilities for Clap:

```python
for t in range(viterbi_post.shape[1] -1):
    vit_b_c2 = np.argmax(T[:,1]*viterbi_post[:,t])
    B[1,t+1] = vit_b_c2
    viterbi_post[1,t+1] = T[vit_b_c2,1]*viterbi_post[vit_b_c2,t]*P[1,t+1]
print(viterbi_post)
```

#### Normalizing the viterbi posterior probability matrix:

```python
vit_norm_1 = np.ones((962,))
vit_norm_2 = np.ones((962,))
for i in range(P.shape[1]):
    vit_norm_1[i] = viterbi_post[0,i]/(viterbi_post[0,i] + viterbi_post[1,i])
    vit_norm_2[i] = viterbi_post[1,i]/(viterbi_post[0,i] + viterbi_post[1,i])
vit_norm = np.vstack((vit_norm_1,vit_norm_2))
print(vit_norm.shape)
```

```python
sns.heatmap(vit_norm)
```

#### Now we will perform back tracking to choose the larger of the two posterior probabilities to select the corresponding B matrix, which would decide the transition state of the frames:

```python
#backtrack_prob = np.zeros((2,962))
#backtrack_prob[:,961] = vit_norm[:,961]
#for i in range(backtrack_prob.shape[1]-1):
   # Bactrack_B = np.argmax(backtrack_prob[:,961-i])
#    f_state = int(B[Bactrack_B, 961-i])
#    backtrack_prob[]
#    backtrack_prob[f_state, 961-i-1] = 1
#    backtrack_prob[1-f_state, 961-i-1] = 0
    
result=[0]*B.shape[1]
result[-1]=np.argmax(vit_norm[:,-1])
for i in range(len(result)-2,-1,-1):
    b=int(result[i+1])
    result[i]=B[b,i]
    
final_result=np.zeros(vit_norm.shape).T
for i in range(len(result)):
    if result[i]:
        final_result[i,1]=1
    else:
        final_result[i,0]=1
final_result = final_result.T
```

#### Plotting the heatmap of the backtracking result of viterbi algorithm:

```python
sns.heatmap(final_result)
```

## Problem 2: Multidimensional Spacing

#### Loading the squared pairwise Euclidean distance matrix:

```python
MDS = loadmat('MDS_pdist.mat')
MDS_L = MDS['L']
print(MDS_L.shape)
```

#### For performing MDS, we first calculate the row average and substract it from the matrix MDS_L:

```python
MDS_L_row = MDS_L - np.mean(MDS_L, axis=0)
```

#### Now, we calculate the column average of the updated matrix and substract it from the matrix MDS_L_row to obtain the matrix W:

```python
W = MDS_L_row - np.mean(MDS_L_row, axis=1)
```

```python
print(W.shape)
```

#### Now we need to divide the above matrix by (-2) in order to obtain a positive definite matrix:

```python
W = W/(-2)
```

#### Performing Eigen Decomposition on the W matrix using SVD as follows:

```python
u,v,_=np.linalg.svd(W)
print(v.shape)
Y_T=np.dot(u,np.sqrt(np.diag(v)))
sns.scatterplot(x=Y_T[:,0],y=Y_T[:,1])
```

#### We can also observe the rotation and shifting of MDS vector by swapping the axis of two largest eigen vectors as follows:

```python
sns.scatterplot(x=Y_T[:,1],y=Y_T[:,0])
```

## Problem 3: Kernel PCA

#### Importing the concentric data set:

```python
concentric = loadmat('concentric.mat')
conc_X = concentric['X']
print(conc_X.shape)
```

#### Plotting a scatter plot:

```python
sns.scatterplot(x=conc_X[0],y=conc_X[1])
```

### Performing the Kernel PCA on this data set:

#### Calculating the matrix Z by applying RBF function using the value:

```python
sigma = 1
kernel_mat = np.exp(-1*distance.pdist(conc_X.T,'sqeuclidean')/np.power(sigma,2))
print(kernel_mat.shape)
kernel_mat= distance.squareform(kernel_mat)
print(kernel_mat.shape)
```

#### Performing eigen decomposition on K:

```python
eig_val_K, eig_vec_K = np.linalg.eigh(kernel_mat)
print(eig_val_K.shape)
print(eig_vec_K.shape)
```

#### Choosing top 3 Eigen vectors and getting the new PCA matrix with a dimension of 3x152:

```python
idx = eig_val_K.argsort()[::-1]   
eig_val_K = eig_val_K[idx]
eig_vec_K = eig_vec_K[:,idx]
PCA_K = np.column_stack([eig_vec_K[:,(2-i)] for i in range(3)])
print(PCA_K.shape)
Z = PCA_K.T
```

#### Calculating the coefficients after PCA:

```python
coef_y = np.dot(PCA_K.T,kernel_mat)
print(coef_y.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
#ax = fig.add_subplot(111,projection = '3d')
ax.scatter3D(Z[0,:],Z[1,:],Z[2,:],c = Z[2,:],cmap = 'RdBu')
plt.show()
```

#### We can see from the 3D scatter plot that the data is linearly separable

#### We also need to label the data samples as 0 and 1. First 51 samples represent the inner circle and the rest as outer circle. Assigning '0' to inner circle and '1' to outer circle data points:

```python
label_zero = np.zeros((51,1))
label_ones = np.ones((101,1))
labels = np.concatenate((label_zero,label_ones),axis=0)
print(labels.shape)
#Z = np.concatenate((Z,labels),axis=0)
#print(Z.shape)
```

#### Training a perceptron to to do the linear classification. First we initialize the three weights and bias:

```python
np.random.seed(200)
weights = np.random.randn(3,1)
bias = np.random.randn(1,1)
print(weights)
print(bias)
```

#### Training the weights and bias using a perceptron model:

```python
def logit(x):
    return 1/(1+np.exp(-x))
epoch = 30000
np.random.shuffle(Z)
learn_rate = 0.01
sse = []
for i in range(epoch):        
        act_input = np.dot(Z.T,weights) + bias
        output_y = logit(act_input)
        error = output_y - labels
        sse.append(np.sum((1/2)*(np.power(error,2))))
        delta = error*output_y*(1-output_y)
        weights = weights - learn_rate*np.dot(Z,delta)
        bias = bias - np.sum(learn_rate*delta)
```

```python
final_output = output_y
predictions =[]
for i in final_output:
    if(i>0.5):
        predictions.append(1)
    else:
        predictions.append(0)
correc_pred = 0
for l in range(len(labels)):
    if (labels[l]==predictions[l]):
        correc_pred = correc_pred + 1
accuracy = (np.sum(correc_pred)/len(labels))*100
print("Accuracy: ",accuracy)
```

```python
plt.plot(sse)
plt.xlabel("Epochs")
plt.ylabel("SSE")
```

#### It can be seen that the sum of squared error(SSE) is minimizing with the increasing number of iterations(epochs)
