# P1- Part 3 Preparing the data matrix for spectral analysis

```python
import math

#The following function process the input audio samples through padding
def process_audio_samples(sample,N):
    quotient=len(sample)//N
    padding_required=round((quotient+1-(len(sample)/N))*N)
    zeros=np.zeros((1,padding_required))
    processed_audio=np.append(sample,zeros)
    return processed_audio
```

```python
X=[]
left_index=0
right_index=N

count=0
hann_window=np.hanning(N)
p_audio=process_audio_samples(audio_samples,N)
while right_index<=len(p_audio):
    audio_slice=audio_samples[left_index:right_index]
    trans_frame=np.multiply(hann_window,p_audio[left_index:right_index])
    X.append(trans_frame)
    left_index+=int(N/2)
    right_index+=int(N/2)
    count+=1
    
X=np.array(X)
```

```python
X=X.T
```

Let's first initialize the basis vector for IDFT.

```python
idft_matrix=[]
for freq in range(0,N):
    ind_comp=[]
    for n in range(0,N):
        ind_comp.append((np.exp(2j*np.pi*(n/N)*freq))*1/N)
    idft_matrix.append(ind_comp)

idft_matrix=np.array(idft_matrix)
```

```python
plt.imshow(idft_matrix.real, aspect='auto')
plt.colorbar()
plt.title('Real component of IDFT Matrix\n');
plt.xlabel('n');
plt.ylabel('frequency');
```

```python
plt.imshow(idft_matrix.imag, aspect='auto')
plt.colorbar()
plt.title('Imaginary component of IDFT Matrix\n');
plt.xlabel('n');
plt.ylabel('frequency');
```

```python
I=dft_matrix.dot(idft_matrix)
plt.imshow(I.real)
plt.colorbar()
plt.title('Real Component of the Identity Matrix\n');
plt.xlabel('n');
plt.ylabel('frequency');
```

Above we can cleary see that the off-diagonal elements have a magnitude of Zero, and the diagonal elements have magnitude close to One. So, we can conclude that our IDFT matrix is valid.

```python
inv_spect=idft_matrix.dot(spect) #Projecting the spectrum on the basis vectors defined by IDFT matrix.
```

```python
X=np.copy(np.real(inv_spect))
r_sig=X[:,0]

for i in range(1,X.shape[1]):
    column=X[:,i]
    ind1=-N//2
    ind2=N//2
    
    r_sig[ind1:]+=X[:ind2,i]
    
    r_sig=np.append(r_sig,X[ind2:,i])
print('Length of recovered signal',len(r_sig))
```

According to the theory of parallax, the stars in our galaxy have larger displacement than the ones away from our galaxy.

<b>Step 4:</b> Implement K-Means to cluster the stars based on their displacement.

```python
import random
def KMeans(data):
    
    #Intialize centroids
    #min_d=min(data)
    #max_d=max(data)
    #centroids=[(3*min_d+max_d)/4, (3*max_d+min_d)/4]
    centroids=random.sample(list(set(data)),2)
    print('Initial Centroids',centroids)
    
    
    
    dist=[]
    for i in range(50):
        #Cluster assignment
        clus1=[]
        clus2=[]
        for j in data:
            distance1=np.linalg.norm(j-centroids[0])
            distance2=np.linalg.norm(j-centroids[1])
            #print(distance1)
            #print(distance2)
            
            if distance1<distance2:
                clus1.append(j)
            else:
                clus2.append(j)
        
        #Update clusters
        centroids[0]=np.mean(clus1)
        centroids[1]=np.mean(clus2)
        
        
        dist1=sum([np.linalg.norm(k-centroids[0]) for k in clus1])
        dist2=sum([np.linalg.norm(l-centroids[1]) for l in clus2])
        print('Epoch-',i+1,'Sum of squared distances in cluster 1',dist1,'Sum of squared distances in cluster 2',dist2,'Cluster centroids',centroids)
        dist.append(dist1+dist2)
    
    plt.plot(dist)
    plt.title('Distance decay')
    plt.xlabel('Epochs')
    plt.ylabel('Distance')
        
    return (clus1,clus2,centroids)

result=KMeans(disp)
cluster1=result[0]
cluster2=result[1]
clus_centroids=result[2]

#print(cluster1)
#print(cluster2)
print(clus_centroids)
```

<b>Step 5:</b> Plot a histogram illustrating the clustering by K-Means.

```python
from matplotlib import pyplot as plt
def plot_histogram(clus_1,clus_2):
    plt.figure(figsize=[10,8]);
    plt.hist(x=clus_1, bins='auto', color='red',alpha=0.7, rwidth=0.85);
    plt.hist(x=clus_2, bins='auto', color='blue',alpha=0.7, rwidth=0.85);
    plt.grid(axis='y', alpha=0.75);
    plt.xlabel('Shift',fontsize=15);
    plt.ylabel('Frequency',fontsize=15);
    plt.xticks(fontsize=10);
    plt.yticks(fontsize=10);
    plt.title('Disparity Histogram',fontsize=15);
    plt.show();
    
plot_histogram(cluster1,cluster2)
```

# P3: GMM for Parallax

<b>Step 1: </b> Let's load the coordinates of the stars as observed in months June and December.

```python
from scipy.io import loadmat
jun=loadmat('june.mat')
dec=loadmat('december.mat')
```

```python
jun=jun['june']
dec=dec['december']
```

<b>Step 2: </b> Calculate the displacement of starts along x-axis.

```python
disp=dec[:,0]-jun[:,0]
```

<b>Step 3: </b> Plot a histogram illustrating the frequency of displacement.

```python
from matplotlib import pyplot as plt
def plot_histogram(dis_arr):
    plt.figure(figsize=[10,8]);
    plt.hist(x=dis_arr, bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85);
    plt.grid(axis='y', alpha=0.75);
    plt.xlabel('Shift',fontsize=15);
    plt.ylabel('Frequency',fontsize=15);
    plt.xticks(fontsize=10);
    plt.yticks(fontsize=10);
    plt.title('Disparity Histogram',fontsize=15);
    plt.show();
    
plot_histogram(disp)
```

<b>Step 4: </b> Let's initialize the parameters for EM algorithm using the result of k-means. 

```python
#From K-Means
import numpy as np
cent1=27.74271844660194
cent2=42.31183368869936
var1=48.17409746441702
var2=19.290286403044174
p1=824/2700 #Probability of cluser 1
p2=1876/200 #Probability of cluster 2

mean=np.array([cent1,cent2])
var=np.array([var1,var2])
P=np.array([p1,p2])
```

<b>Step 5: </b> Write the functions to update the parameters of a GMM using the equations from lecture slides.

```python
import numpy as np
def gauss_pdf(x,mean,var):
    in_exp=-1/2*(x-mean)*(1/var)*(x-mean)
    return ((1/(2*np.pi*var)**0.5)*np.exp(in_exp))
    

def update_U(P,mean,var,x):
    U_1=gauss_pdf(x,mean[0],var[0])
    U_2=gauss_pdf(x,mean[1],var[1])
    U=np.column_stack((U_1,U_2))
    
    for i in range(0,len(U)):
        denom=(P[0]*U[i,0]+P[1]*U[i,1])
        num1=P[0]*U[i,0]
        num2=P[1]*U[i,1]
        U[i,0]=num1/denom
        U[i,1]=num2/denom
    
    return U
    
```

```python
def update_mean(U,x):
    U_1=U[:,0]
    U_2=U[:,1]
    sum1=sum(U_1)
    sum2=sum(U_2)
    mean1=(U_1.dot(x))/sum1
    mean2=(U_2.dot(x))/sum2
    
    return np.array([mean1,mean2])

def update_cluster_prob(U):
    U_1=U[:,0]
    U_2=U[:,1]
    P1=sum(U_1)/len(U_1)
    P2=sum(U_2)/len(U_2)
    return np.array([P1,P2])

def update_covar(U,x,mean):
    U_1=U[:,0]
    U_2=U[:,1]
    mean1=mean[0]
    mean2=mean[1]
    covar1=U_1.dot((x-mean1)*(x-mean1))/sum(U_1)
    covar2=U_2.dot((x-mean2)*(x-mean2))/sum(U_2)
    
    return np.array([covar1,covar2])


    
```

<b>Step 6: </b> Update the parameters of the GMM uing the EM algorithm for 500 iterations. 

```python
for i in range(500):
    
    #E-step
    U=update_U(P,mean,var,disp)
    
    #M-step
    mean=update_mean(U,disp)
    
    P=update_cluster_prob(U)
    
    var=update_covar(U,disp,mean)
    
    
```

```python
print('Cluster means',mean)
print('Cluster probabilities',P)
print('Cluser variances',var)
```

```python
max_index=[]
for i in U:
    max_index.append(np.argmax(i))
```

<b> Step 7: </b> Segregate the data points into clusters based on the results of EM.

```python
clus_1=[]
clus_2=[]
for i in range (0,len(disp)):
    if max_index[i]:
        clus_1.append(disp[i])
    else:
        clus_2.append(disp[i])
```

```python
print(len(clus_1))
print(len(clus_2))
```

<b> Step 8: </b> Segregate the stars as observed in June into clusters based on the results of EM.

```python
jun_clus1=[]
jun_clus2=[]
for i in range(0,len(max_index)):
    if max_index[i]:
        jun_clus1.append(jun[i])
    else:
        jun_clus2.append(jun[i])

jun_clus1=np.array(jun_clus1)
jun_clus2=np.array(jun_clus2)       
```

```python
fig, ax = plt.subplots()

ax.scatter(jun_clus1[:,0],jun_clus1[:,1],c='tab:red',label='Our Galaxy',alpha=0.5, edgecolors='none')
ax.scatter(jun_clus2[:,0],jun_clus2[:,1],c='tab:blue',label='Other Galaxy',alpha=0.5, edgecolors='none')
ax.legend()
ax.grid(True)

plt.show()
```

<b> Step 9: </b> Plot a histogram to visualize the frequency distribution in each cluster.

```python
from matplotlib import pyplot as plt
def plot_histogram(clus_1,clus_2):
    plt.figure(figsize=[10,8])
    plt.hist(x=clus_1, bins='auto', color='red',alpha=0.7, rwidth=0.85,label='Our galaxy')
    plt.hist(x=clus_2, bins='auto', color='blue',alpha=0.7, rwidth=0.85,label='Other galaxy')
    plt.grid(axis='y', alpha=0.75)
    plt.xlabel('Shift',fontsize=15)
    plt.ylabel('Frequency',fontsize=15)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.title('Disparity Histogram',fontsize=15)
    plt.legend();
    plt.show();
    
    
plot_histogram(clus_1,clus_2)
```

<b> Conclusion: </b> I would prefer the GMM over K-means because GMM makes a soft decision for attributing data points to clusters by taking into count the probability and covariance of clusters. Also, unlike K-Means, segregation achieved through GMM agrees with Mahalanobis distance of data points from each cluster centroids. This soft decision characteristic is evident in the results of our models. For example: Our KMeans models has cluster centroids at- (27.74271844660194, 42.31183368869936), and variances (48.17409746441702, 19.290286403044174). Now, a sample value of 35.5 would be equidistant from both the cluster means in our k-means model and grouped in cluster 2. But the GMM model groups such point in cluster 1, as cluster 1 has higher variance. Therefore, the GMM model makes a sooft decision by abiding by the underlying distribution of data points.

# P4: DCT and PCA

```python
import matplotlib.image as mpimg 
import random
```

<b>Step 2 :</b> Divide the color image into three channels.

```python
Xr=img[:,:,0]
Xg=img[:,:,1]
Xb=img[:,:,2]
```

```python
print('Shape of red channel',Xr.shape)
print('Shape of green channel',Xg.shape)
print('Shape of blue channel',Xb.shape)
```

<b>Step 3:</b> Write a function which randomly chooses rows from the input channels and creates a 8 dimensional block by concatenating channels.   

```python
import numpy as np
def extract_channels(Xr,Xg,Xb, blocks):
    R=np.array(Xr)
    G=np.array(Xg)
    B=np.array(Xb)
    
    r_indexes=random.sample(list(range(0,R.shape[0]-8)),blocks//3)
    g_indexes=random.sample(list(range(0,G.shape[0]-8)),blocks//3)
    b_indexes=random.sample(list(range(0,B.shape[0]-8)),blocks//3)
    
    
    r_channels=np.array([R[i:i+8,:] for i in r_indexes])
    b_channels=np.array([B[i:i+8,:] for i in b_indexes])
    g_channels=np.array([G[i:i+8,:] for i in g_indexes])
    

    r_block=r_channels[0]
    for i in range(1,r_channels.shape[0]):
        r_block=np.concatenate((r_block,r_channels[i]),axis=1)
    
    
    
    b_block=b_channels[0]
    for i in range(1,b_channels.shape[0]):
        b_block=np.concatenate((b_block,b_channels[i]),axis=1)
    
    
    
    g_block=g_channels[0]
    for i in range(1,g_channels.shape[0]):
        g_block=np.concatenate((g_block,g_channels[i]),axis=1)
    
    
    
    
    final_block=np.concatenate((r_block,g_block,b_block),axis=1)
    print('Dimension of the R block is',final_block.shape)
    
    return final_block
```

<b>Step 4: </b> Create a (8 X 6192) block by concatenating 3 8-dimensional blocks from each channel (9 in total). 

```python
R=extract_channels(Xr,Xg,Xb,9)
R=R.astype('float64')
```

<b>Step 5: </b> Center the data points by subtracting the mean. 

```python
for i in range(0,len(R)):
    mean=np.mean(R[i])
    R[i]-=mean
```

<b>Step 6: </b> Calculate the dot product by multiplying the data-matrix with its transpose.

```python
co_var=R.dot(R.T)
```

```python
co_var.shape
```

<b>Step 7: </b> Calculate the eigen vectors using linalg module of numpy.

```python
eig_val_1,eig_vect_1=np.linalg.eig(co_var)
```

```python
eig_val_1
```

Above we can see that the eigen values are ordered in decreasing order.

```python
eig_vect_1
```

<b>Step 9: </b> Create a (8 X 61920) block by concatenating 30 8-dimensional blocks from each channel (90 in total) and repeat steps 5-8 .

```python
for i in range(0,len(R)):
    mean=np.mean(R[i])
    R[i]-=mean
```

<b>Step 10: </b> Plot the DCT matrix.

```python
dct_matrix=[]
for k in range(0,8):
    ind_comp=[]
    for n in range(0,8):
        ind_comp.append(np.sqrt(1/8)*np.cos(np.pi*(2*n+1)*k/16))
    
    dct_matrix.append(ind_comp)
```

```python
dct_matrix=np.array(dct_matrix)
dct_matrix.shape
```

Since the difference between DCT matrix and second set of eigen vectors is lower than the difference between DCT matrix and first set of eigen vectors, the second set of eigen vectors have more resemblance with the DCT matrix.

<b> Conclusion: </b> Above we can see that the DCT and Eigen vectors matrices are similar with some marginal difference in magnitude. Additionally, we have observed that the eigen vectors matrix generated using a block size of 90 has more resemblance to DCT matrix than the eigen vectors matrix generated using a block size of 9.Therefore, the quality of transformation through PCA can be improved by increasing the size of our data set (R). .
