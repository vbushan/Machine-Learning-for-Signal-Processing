```python
from matplotlib import pyplot as plt
def plot_histogram(dis_arr):
    plt.figure(figsize=[10,8])
    plt.hist(x=dis_arr, bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85)
    plt.grid(axis='y', alpha=0.75)
    plt.xlabel('Shift',fontsize=15)
    plt.ylabel('Frequency',fontsize=15)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.title('Disparity Histogram',fontsize=15)
    plt.show();
    
plot_histogram(disparity)
```

```python
def cent_update(data,lables):
    cls_1=[]
    cls_2=[]
    for i,j in zip(data,lables):
        if j==0:
            cls_1.append(i)
        else:
            cls_2.append(i)
    
    print(len(cls_2))
    return (np.mean(cls_1),np.mean(cls_2))


int_labels=clust_assign(disparity,int_cents)
print(cent_update(disparity,int_labels))
            
```

```python
import random
def kMeans_i(data,K):
    init_centroids=random.sample(list(data),K)
    print(init_centroids)
    
    
    #cl_labels=[]
    
    cents=list(init_centroids)
    iter=100
    for i in range(0,iter):
        
        
        
        cl_labels=[]
        
        #Cluster Assignment
        for j in range(0,len(data)):
            cl_labels.append(np.argmin([(data[i]-cents[0])**2,(data[i]-cents[1])**2]))
            
        
        #Update Centroid
        cls1=[]
        cls2=[]
        for k in range(0,len(cl_labels)):
            if cl_labels[k]==0:
                cls1.append(data[k])
            else:
                cls2.append(data[k])
        
        cents.clear()
        cents.append(np.mean(cls1))
        cents.append(np.mean(cls2))
        
        
    return cl_labels
    
    
        
        
    
print(kMeans_i(disparity,2))
```
