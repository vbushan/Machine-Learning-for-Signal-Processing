```python
import numpy as np
disparity_matrix=[]
for i in range(X_L.shape[0]):
    temp=[]
    for j in range(X_L.shape[1]-40):
        #if not(j>X_L.shape[1]-40):
        #    window=X_L[i,-40:,:]
        #else:
        window=X_L[i,j:j+40,:]
        
        similarity=window-X_R[i,j,:]
        
        similarity=similarity*similarity
        
        similarity=np.sum(similarity,axis=1)
        
        similarity=list(similarity)
        similarity=[(k,similarity[k]) for k in range(len(similarity))]
        similarity.sort(key=lambda x: x[1])
        index=similarity[0][0]
        temp.append(index)
    
    disparity_matrix.append(temp)
    
        
```

```python
#GMM
#K Means
import random
def KMeans(data,n_cents):
    
    #Intialize centroids
    
    centroids=random.sample(list(set(data)),n_cents)
    print('Initial Centroids',centroids)
    
    
    
    tot_dist=[]
    for i in range(50):
        clus={i:[] for i in range(n_cents)}            
        #Cluster assignment
        
        dist=centroids-data.reshape(-1,1)
        dist=dist*dist
        
        temp=np.argmin(dist,axis=1)
        for j in range(len(temp)):
            clus[temp[j]]+=[data[j]]
        
        
        #Update clusters
        for l in range(len(centroids)):
            centroids[l]=np.mean(clus[l])
            
        
        #Calculate total distance
        temp_dist=0
        for m in range(len(centroids)):
            temp_dist+=sum([np.linalg.norm(n-centroids[m]) for n in clus[m]])
        
        tot_dist.append(temp_dist)
    
    plt.plot(tot_dist)
    plt.title('Distance decay')
    plt.xlabel('Epochs')
    plt.ylabel('Distance')
        
    return (tuple(clus.values()),centroids)
```

```python
clusters,cents=KMeans(disparity_matrix,4)
```

```python
import numpy as np
def gauss_pdf(x,mean,var):
    in_exp=-1/2*(x-mean)*(1/var)*(x-mean)
    return ((1/(2*np.pi*var)**0.5)*np.exp(in_exp))
    

def update_U(P,mean,var,x):
    temp_U=[]
    for i in range(len(mean)):
        temp_U.append(gauss_pdf(x,mean[i],var[i]))
    
    U=np.column_stack(tuple(temp_U))
   
    U=U*P/(U.dot(P.reshape(-1,1)))
    
    
    return U

```

```python
def update_mean(U,x):

    #temp_mean=[]
    #for i in range(U.shape[1]):
    #    temp_mean.append((U[:,i].dot(x))/sum(U[:,i]))
    
    temp_mean=np.sum(U*x.reshape(-1,1),axis=0)/np.sum(U,axis=0)
    
    return temp_mean

def update_cluster_prob(U):

    
    #temp_prob=[]
    #for i in range(U.shape[1]):
    #    temp_prob.append(sum(U[:,i])/len(U[:,1]))
    
    temp_prob=np.sum(U,axis=0)/len(U)
    
    return temp_prob

def update_covar(U,x,mean):

    
    temp_covar=[]
    for i in range(U.shape[1]):
        temp_covar.append(U[:,i].dot((x-mean[i])*(x-mean[i]))/sum(U[:,i]))
    
    

    return np.array(temp_covar)

```

```python
for i in range(100):
    
    #E-step
    U=update_U(P,mean,var,disparity_matrix)
    
    #M-step
    mean=update_mean(U,disparity_matrix)
    
    P=update_cluster_prob(U)
    
    var=update_covar(U,disparity_matrix,mean)
    
    
```
